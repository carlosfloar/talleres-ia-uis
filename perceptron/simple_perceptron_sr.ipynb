```python
from google.colab import output
output.enable_custom_widget_manager()
%pip install pillow
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from PIL import Image
from io import BytesIO
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.lines import Line2D
import requests
np.random.seed(42)

# Base Imgur IDs of the images
orange_imgur_id = "J8NwANo"
banana_imgur_id = "SxuqIcZ"

# Function to download image from Imgur ID and return as PIL Image
def load_image_from_imgur_id(imgur_id, size=(30, 30)):
    # Construct the direct image URL from the Imgur ID
    direct_image_url = f"https://i.imgur.com/{imgur_id}.png"

    # Add a User-Agent header to mimic a browser request
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    response = requests.get(direct_image_url, headers=headers)
    # Check if the request was successful
    response.raise_for_status() # Raise an exception for HTTP errors
    img = Image.open(BytesIO(response.content))
    if img.mode != 'RGBA':
        img = img.convert('RGBA')
    img = img.resize(size, Image.Resampling.LANCZOS)
    return img

# Load and resize images using Imgur IDs
orange_img = load_image_from_imgur_id(orange_imgur_id)
bana_img = load_image_from_imgur_id(banana_imgur_id)

N = 20

# Clase 0
X0 = np.random.randn(N, 2) + np.array([-2, -2])

# Clase 1
X1 = np.random.randn(N, 2) + np.array([2, 2])

X = np.vstack([X0, X1])
y = np.array([0]*N + [1]*N)

def plot_decision_boundary(w1, w2, b):
    fig, ax = plt.subplots(figsize=(5,5))

    # Dibujar puntos como imágenes
    for x_val, y_label in zip(X, y):
        img_to_use = orange_img if y_label == 0 else bana_img
        ab = AnnotationBbox(OffsetImage(img_to_use), (x_val[0], x_val[1]), frameon=False, pad=0.1)
        ax.add_artist(ab)

    # Frontera de decisión
    x_vals = np.linspace(-6, 6, 200)
    if w2 != 0:
        y_vals = -(w1 * x_vals + b) / w2
        ax.plot(x_vals, y_vals, linewidth=2, color='red', linestyle='--')
    else:
        # Manejar el caso de línea vertical si w2 es 0
        if w1 != 0:
            ax.axvline(x=-b/w1, color='red', linestyle='--', linewidth=2)

    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
    ax.set_title(f"w = ({w1:.2f}, {w2:.2f})   b = {b:.2f}")

    # Crear etiquetas personalizadas para la leyenda de las imágenes
    custom_lines = [
        Line2D([0], [0], marker='o', color='w', label='Clase 0 (Naranja)', markerfacecolor='orange', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Clase 1 (Plátano)', markerfacecolor='yellow', markersize=10)
    ]
    ax.legend(handles=custom_lines, loc='upper left')

    ax.grid(True)
    plt.show()
```

```python
interact(
    plot_decision_boundary,
    w1=FloatSlider(min=-1, max=1, step=0.1, value=1),
    w2=FloatSlider(min=-1, max=1, step=0.1, value=1),
    b=FloatSlider(min=-5, max=5, step=0.1, value=0)
);
```
